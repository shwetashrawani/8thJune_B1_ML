```python
def data_impute(x):
    if pd.isnull(x['Age']):
        return titanic[(titanic['Sex']==x['Sex'])&(titanic['Pclass']==x['Pclass'])]['Age'].mean()
    return x['Age']
```

```python
titanic['Age']=titanic[['Sex','Pclass','Age']].apply(data_impute,axis=1)
```

```python
titanic.dropna()                                   # dropna is used to remove null values
```

```python
titanic.dropna(inplace=True)
```

```python
titanic['Sex'].str.get_dummies()                                #b convert string into integer
```

```python
titanic_1=pd.concat([titanic['Sex'].str.get_dummies()],axis=1)
titanic_1.head()
```

```python
titanic_1=pd.concat([titanic,titanic['Sex'].str.get_dummies()],axis=1)
titanic_1.head()
```

```python
x=titanic_1.drop(['Survived','Embarked','Sex','Ticket'],axis=1)
y=titanic_1['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt=DecisionTreeClassifier().fit(xtrain,ytrain)
```

```python
print(round(model_dt.score(xtest,ytest)*100,2))
```
