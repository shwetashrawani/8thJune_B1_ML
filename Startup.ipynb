```python
y = x - 3*(x**2) + 0.6*(x**3) + np.random.normal(0,1,30)         #generating 30 polynomial no bte 0 and 1
y
```

```python
y.shape
```

```python
plt.scatter(x,y,s=25,c='b')
```

```python
x = x[:,np.newaxis]         # adding new axis
x.shape
```

```python
y = y[:,np.newaxis]
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression().fit(x,y)
```

```python
ypred = model1.predict(x)     # predicting y on the basis of x
```

```python
model1.score(x,y)*100
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.plot(x,ypred,c='b')
plt.grid()
```

```python
from sklearn import preprocessing as pp     # preprocessing is use to convert one data into multideminsions.
```

```python
x.shape
```

```python
x
```

```python
x_poly1 = pp.PolynomialFeatures(1).fit_transform(x)   #pp.PolynomialFeatures(1) here we use (1) so it will convert data into 2-D array
x_poly1
```

```python
x_poly1 = pp.PolynomialFeatures(1).fit_transform(x)   #we are converting x which is random data into polynomial data like y which is alredy in polynomial format.
model_d1 = LinearRegression().fit(x_poly1,y)          #we are doing changes only in model data (x).
ypred_d1 = model_d1.predict(x_poly1)
print(100*model_d1.score(x_poly1,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.plot(x,ypred_d1,c='b')
plt.grid()                    # here in this graph blue line represents our prediction model(convesion x random data to polynomial) and red scatterplot shows us actual data(x=random data,y=polynomial data.) 
```

```python
x_poly2 = pp.PolynomialFeatures(2).fit_transform(x)   #by increasing polynomialfeatures degree from 1 to 2 our prediction data (x) slightly improve into polynomial. 
model_d2 = LinearRegression().fit(x_poly2,y)
ypred_d2 = model_d2.predict(x_poly2)
print(100*model_d2.score(x_poly2,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
# plt.plot(x,ypred_d2,c='b')
plt.scatter(x,ypred_d2,c='b')
plt.grid()
```

```python
x_poly3 = pp.PolynomialFeatures(3).fit_transform(x)
model_d3 = LinearRegression().fit(x_poly3,y)
ypred_d3 = model_d3.predict(x_poly3)
print(100*model_d3.score(x_poly3,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
# plt.plot(x,ypred_d2,c='b')
plt.scatter(x,ypred_d3, c='b')
plt.grid()
```

```python
x_poly4 = pp.PolynomialFeatures(4).fit_transform(x)
model_d4 = LinearRegression().fit(x_poly4,y)
ypred_d4 = model_d4.predict(x_poly4)
print(100*model_d4.score(x_poly4,y))
plt.figure(figsize=(10,6))
plt.scatter(x,y,s=25,c='r')
plt.scatter(x,ypred_d4, c='b')
plt.grid()                          # Our x properly convert into polynomial data
```

```python
score = []
for i in range(1,20):
    x_poly_i = pp.PolynomialFeatures(i).fit_transform(x)
    model_d_i = LinearRegression().fit(x_poly_i,y)
    score.append(100*model_d_i.score(x_poly_i,y))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,20),score,marker='o',)
plt.grid()
```
